# 4. Data Cleaning

# Price & Discount Columns Need Numeric Conversion

```python
# Price & Discount Columns Need Numeric Conversion

data['discounted_price'] = data['discounted_price'].astype(str).str.replace('₹','').str.replace(',','')

data['actual_price'] = data['actual_price'].astype(str).str.replace('₹','').str.replace(',','')

data['discount_percentage'] = data['discount_percentage'].astype(str).str.replace('%','')

# Converting to numeric float format

data['discounted_price'] = data['discounted_price'].astype(float)
data['actual_price'] = data['actual_price'].astype(float)
data['discount_percentage'] = data['discount_percentage'].astype(float)
data.head()
```

# Rating & Rating Count Require Fixing

```python
# Removing (,) form rating_count
data['rating_count'] = data['rating_count'].astype(str).str.replace(',','')


# Convert to numeric first, which will properly handle 'nan' strings
data['rating_count'] = pd.to_numeric(data['rating_count'],errors='coerce')

# filling NaN with 0

data['rating_count'] = data['rating_count'].fillna(0)

# Converting to integer value

data['rating_count'] = data['rating_count'].astype(int)

data['rating'] = pd.to_numeric(data['rating'], errors='coerce')

# Fill NaN values if needed
data['rating'] = data['rating'].fillna(0)  # or another appropriate value

# Converting to float value
data['rating'] = data['rating'].astype(float)


data.info()

data.isnull().sum()
```

# Category Field Is Long & Fragmented


```python
# Extract only the first category

data['category'] = data['category'].astype(str).str.split('|').str[0]

#Remove extra spaces if present

data['category'] = data['category'].str.strip()

data.head()
```

```python
# Removing the exta space from each columns

data['product_name'] = data['product_name'].str.strip()
data['about_product'] = data['about_product'].str.strip()
data['review_title'] = data['review_title'].str.strip()
data['review_content'] = data['review_content'].str.strip()
```

# 5. Feature Engineering 

```python
# exracting the discount amount 

data['discount_amount'] = data['actual_price'] - data['discounted_price']

# extracting high discount products by assigning value 1 for high and 0 for low

data['high_discount'] = data['discount_percentage'].apply(lambda x: 1 if x >= 40 else 0)


data.head()
```

```python
# Extacting price range Bucketed ranges (0–500, 500–1000, etc.)
bins = [0,500,1000,1500,2000,3000,5000,10000,float('inf')]

labels = [
    '0-500',
    '500-1000',
    '1000-1500',
    '1500-2000',
    '2000-3000',
    '3000-5000',
    '5000-10000',
    '10000+'
]

    
data['price_range'] = pd.cut(data['discounted_price'], bins=bins , labels=labels , right=False)

data.head()
```

```python
# Rating buckets Group ratings into 4 categories (<2, 2–3, 3–4, 4–5)
bins = [0,2,3,4,5]
labels = [
    '<2',
    '2-3',
    '3-4',
    '4-5'
]

data['rating_buckets'] = pd.cut(data['rating'] , bins=bins , labels=labels , right=False)

# Exracting net saving percent 

data['net_saving_percent'] = round(data['discounted_price'] / data['actual_price'] * 100,2)
data.head()
```

# DESIGN SQL TABLES (Schema + ERD)

```python
import sqlite3 as sp

connt = sp.connect('amazon.db')
cursor = connt.cursor()
```

```python
cursor.executescript("""CREATE TABLE IF NOT EXISTS products(
product_id TEXT PRIMARY KEY,
product_name TEXT,
category TEXT,
discounted_price REAL,
actual_price REAL,
discount_percentage REAL,
img_link TEXT,
product_link TEXT);

CREATE TABLE IF NOT EXISTS reviews(
review_id TEXT PRIMARY KEY,
product_id TEXT,
user_id TEXT,
user_name TEXT,
rating REAL,
rating_count INTEGER,
review_title TEXT,
review_content TEXT,
FOREIGN KEY (product_id) REFERENCES products(product_id)
);
""")

connt.commit()
connt.close()
```

# PREPARE DATA TO INSERT

```python
# Perpareing a data set for product table to insert data from dataset

products_df = data[[
    'product_id','product_name','category',
    'discounted_price','actual_price','discount_percentage',
    'img_link','product_link'
]].drop_duplicates(subset='product_id')
```

```python
## Perpareing a data set for review table table to insert data from dataset

reviews_df = data[[
    'review_id','product_id','user_id','user_name',
    'rating','rating_count','review_title','review_content'
]]
```

```python
#INSERT DATA INTO SQLITE

connt = sp.connect('amazon.db')
cursor = connt.cursor()

products_df.to_sql('products', connt, if_exists='replace', index=False)
reviews_df.to_sql('reviews', connt, if_exists='replace', index=False)

connt.commit()
```

# TOP CATEGORIES BY REVIEW VOLUME

```python
# TOP CATEGORIES BY REVIEW VOLUME

query = """
SELECT
p.category,
SUM(r.rating_count) AS total_reviews
FROM products p
JOIN reviews r
ON p.product_id = r.product_id
GROUP BY p.category
ORDER BY total_reviews DESC;
"""

pd.read_sql(query,connt)
```

# Top Categories by Review Volume 

Using SQL, I calculated the total number of customer reviews for each product category.

Electronics emerges as the most dominant category, followed by Computers & Accessories and Home & Kitchen, indicating extremely high customer engagement in these segments.

These categories represent highly competitive markets on Amazon, where products receive significant customer attention and feedback.

Categories such as Office Products and Musical Instruments show comparatively lower review volumes, reflecting niche demand and fewer active products.

# Best-rated categories (with minimum reviews)

```python
# Best-rated categories (with minimum reviews)

query1 = """
SELECT p.category,
ROUND(AVG(r.rating),2) AS avg_rating,
SUM(r.rating_count) AS total_reviews

FROM products p
JOIN reviews r
ON p.product_id = r.product_id
WHERE r.rating IS NOT NULL
GROUP by p.category
HAVING SUM(r.rating_count) >= 100
ORDER by avg_rating DESC;
"""

pd.read_sql(query1,connt)
```

# Best-Rated Categories 

Using SQL, I calculated the average customer rating for each category by joining product and review data.
To ensure reliability, only categories with a minimum review volume were considered.

Office Products, Toys & Games, and Home Improvement emerge as the highest-rated categories, indicating strong customer satisfaction and consistent product quality.

High-volume categories such as Electronics and Computers & Accessories maintain solid average ratings above 4.0 despite intense competition and large review counts.

This suggests that niche categories often deliver higher satisfaction, while mass-market categories balance scale with quality.

# Top 10 Highest Discount Products

```python
# Top 10 Highest Discount Products

query3 = """
SELECT p.product_id,
p.product_name,
p.category,
p.discount_percentage,
p.discounted_price,
p.actual_price,
ROUND(AVG(r.rating),2) AS avg_rating,
SUM(r.rating_count) AS total_reviews
FROM products p
JOIN reviews r
ON p.product_id = r.product_id
GROUP BY p.product_id,
p.product_name,
p.category,
p.discount_percentage,
p.discounted_price,
p.actual_price
ORDER by p.discount_percentage DESC
LIMIT 20;
"""

pd.read_sql(query3,connt)
```

# Highest Discount Products 

Using SQL, I identified the top products offering the highest discounts.

The results show extremely high discount percentages (often above 80%), particularly in Electronics and Computer Accessories categories.

Several highly discounted products still maintain solid average ratings above 4.0, indicating that aggressive discounting is sometimes used as a competitive pricing strategy rather than to compensate for poor quality.

However, a few products combine very high discounts with moderate ratings, suggesting potential issues such as inflated original prices or inconsistent product quality.

# Bad Quality Products

```python
query4 = """
SELECT p.product_id,
p.product_name,
p.category,
p.discount_percentage,
ROUND(AVG(r.rating),2) AS avg_rating,
SUM(r.rating_count) AS total_reviews
FROM products p
JOIN reviews r
ON p.product_id = r.product_id
GROUP BY p.product_id,
p.product_name,
p.category,
p.discount_percentage
HAVING
p.discount_percentage >= 70
AND AVG(r.rating)  <= 3.5
AND SUM(r.rating_count) >= 50
ORDER by p.discount_percentage DESC,
avg_rating ASC;
"""

pd.read_sql(query4,connt)
```

# Bad Quality Products 

Using SQL, I identified products that combine very high discount percentages with low average customer ratings, while also having sufficient review volume to ensure reliability.

The results reveal several products—primarily in Electronics and Computer Accessories—that are discounted by more than 70% yet still receive poor ratings below 3.5.

This pattern suggests that heavy discounting is being used to compensate for weak product quality or customer dissatisfaction.

These products represent high-risk items for customers and highlight areas where sellers should focus on improving product quality rather than relying on aggressive pricing strategies.
